```python
img2 = cv2.imread('comp01.jpeg')
plt.imshow(img2)
```

```python
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
val, thresh_simples = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
plt.imshow(thresh_simples, cmap='gray')
```

```python
val, otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
plt.imshow(otsu, cmap='gray')
print(val)
```

```python
texto = pytesseract.image_to_string(otsu, lang='por')
print(texto)
```

```python
adapt_media = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11,9)
plt.imshow(adapt_media, cmap='gray')
```

```python
texto = pytesseract.image_to_string(adapt_media, lang='por')
print(texto)
```

```python
adapt_gaus = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11,10)
plt.imshow(adapt_gaus, cmap='gray')
```

```python
plt.imshow(255 - adapt_gaus, cmap='gray')
```

```python
texto = pytesseract.image_to_string(255 - adapt_gaus, lang='por')
print(texto)
```

```python
texto = pytesseract.image_to_string(adapt_gaus, lang='por')
print(texto)
```
