```python
import cv2
import pytesseract

def preprocessamento_leve(image_path):
    img = cv2.imread(image_path)
    
    # 1. Aumentar a escala (DPI)
    # O Tesseract performa muito melhor com caracteres grandes. 
    # Dobrar o tamanho com interpolação CUBIC suaviza as letras.
    img = cv2.resize(img, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)
    
    # 2. Converter para Tons de Cinza (sem binarizar ainda)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 3. Leve desfoque (Blur) 
    # Ajuda a conectar os pontos de impressoras térmicas/matriciais
    # que costumam falhar se binarizarmos direto.
    blur = cv2.GaussianBlur(gray, (3,3), 0)
    
    return blur

# Teste com a configuração que você usou
imagem_melhorada = preprocessamento_leve('comp01.jpeg')
texto = pytesseract.image_to_string(imagem_melhorada, lang='por', config=config)
print(texto)
```

```python
import easyocr
import cv2
import re

def extrair_dados_comprovante(caminho_imagem):
    # 1. Carregar a imagem e converter para escala de cinza (melhora o OCR)
    img = cv2.imread(caminho_imagem)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 2. Inicializar o leitor EasyOCR para Português
    reader = easyocr.Reader(['pt'])
    
    # 3. Ler o texto da imagem
    resultados = reader.readtext(gray, detail=0) # detail=0 retorna apenas o texto
    texto_completo = "\n".join(resultados)

    print("--- Texto Extraído ---")
    print(texto_completo)
    print("----------------------\n")

    # 4. Extração de dados específicos usando Regex
    dados = {
        "Estabelecimento": resultados[4] if len(resultados) > 4 else "Não encontrado",
        "Valor": re.search(r"R\$\s?(\d+,\d{2})", texto_completo),
        "Data": re.search(r"(\d{2}/\d{2}/\d{4})", texto_completo),
        "CNPJ": re.search(r"\d{2}\.\d{3}\.\d{3}/\d{4}-\d{2}", texto_completo)
    }

    # Formatação dos resultados capturados pelo Regex
    print("--- Dados Estruturados ---")
    print(f"Loja: {dados['Estabelecimento']}")
    print(f"Valor: {dados['Valor'].group(0) if dados['Valor'] else 'Não encontrado'}")
    print(f"Data: {dados['Data'].group(0) if dados['Data'] else 'Não encontrado'}")
    print(f"CNPJ: {dados['CNPJ'].group(0) if dados['CNPJ'] else 'Não encontrado'}")

# Executar a função
extrair_dados_comprovante('comp02.jpeg')
```

```python
import cv2
import numpy as np
import easyocr
import re

def ordenar_pontos(pts):
    rect = np.zeros((4, 2), dtype="float32")
    s = pts.sum(axis=1)
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]
    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]
    rect[3] = pts[np.argmax(diff)]
    return rect

def transformar_perspectiva(image, pts):
    rect = ordenar_pontos(pts)
    (tl, tr, br, bl) = rect
    widthA = np.sqrt(((br[0] - bl[0]) ** 2) + ((br[1] - bl[1]) ** 2))
    widthB = np.sqrt(((tr[0] - tl[0]) ** 2) + ((tr[1] - tl[1]) ** 2))
    maxWidth = max(int(widthA), int(widthB))
    heightA = np.sqrt(((tr[0] - br[0]) ** 2) + ((tr[1] - br[1]) ** 2))
    heightB = np.sqrt(((tl[0] - bl[0]) ** 2) + ((tl[1] - bl[1]) ** 2))
    maxHeight = max(int(heightA), int(heightB))
    dst = np.array([[0, 0], [maxWidth - 1, 0], [maxWidth - 1, maxHeight - 1], [0, maxHeight - 1]], dtype="float32")
    M = cv2.getPerspectiveTransform(rect, dst)
    return cv2.warpPerspective(image, M, (maxWidth, maxHeight))

def pipeline_final(caminho_imagem):
    # 1. CARREGAMENTO E DETECÇÃO DE BORDAS
    image = cv2.imread(caminho_imagem)
    ratio = image.shape[0] / 800.0
    orig = image.copy()
    res_img = cv2.resize(image, (int(image.shape[1] / ratio), 800))
    
    gray = cv2.cvtColor(res_img, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (5, 5), 0)
    edged = cv2.Canny(gray, 75, 200)

    cnts, _ = cv2.findContours(edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    cnts = sorted(cnts, key=cv2.contourArea, reverse=True)[:5]
    
    warped = orig
    for c in cnts:
        peri = cv2.arcLength(c, True)
        approx = cv2.approxPolyDP(c, 0.02 * peri, True)
        if len(approx) == 4:
            warped = transformar_perspectiva(orig, approx.reshape(4, 2) * ratio)
            break

    # 2. MELHORIA DE QUALIDADE (Rescale + Sharpness)
    # Aumentamos a imagem para o OCR ler melhor as letras miúdas
    warped = cv2.resize(warped, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)
    gray_warped = cv2.cvtColor(warped, cv2.COLOR_BGR2GRAY)
    
    # Filtro Sharpness para destacar o texto
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    processed_img = cv2.filter2D(gray_warped, -1, kernel)

    # 5. Plotagem das Imagens
    plt.figure(figsize=(15, 7))
    
    plt.subplot(1, 2, 1)
    plt.title("Imagem Original")
    plt.imshow(image)
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.title("Processada para OCR (Perspectiva + Nitidez)")
    plt.imshow(processed_img, cmap='gray')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    # 3. OCR
    print("Iniciando leitura OCR...")
    reader = easyocr.Reader(['pt', 'en'])
    resultados = reader.readtext(processed_img, detail=0)
    texto_completo = " ".join(resultados).upper()

    # 4. EXTRAÇÃO COM REGEX REFINADO
    # Valor: Prioriza números com vírgula (R$ 50,00)
    re_valor = re.search(r"(?:R\$|RS|[\s])\s?(\d{1,5},\d{2})", texto_completo)
    
    # Data: Aceita 04/FEV/2026 ou 04/02/2026
    re_data = re.search(r"(\d{2}[/\s](?:[A-Z]{3}|\d{2})[/\s]\d{4})", texto_completo)
    
    # CNPJ: Padrão brasileiro
    re_cnpj = re.search(r"(\d{2}\.?\d{3}\.?\d{3}/?\d{4}-?\d{2})", texto_completo)

    # Estabelecimento: Tenta pegar o texto antes do endereço "RUA"
    estabelecimento = "NÃO ENCONTRADO"
    for i, linha in enumerate(resultados):
        if "RUA" in linha.upper():
            estabelecimento = resultados[i-1] if i > 0 else "POSTO ROCHA"
            break

    # 5. EXIBIÇÃO DOS RESULTADOS
    print("\n" + "="*30)
    print(f"LOJA:  {estabelecimento}")
    print(f"VALOR: R$ {re_valor.group(1) if re_valor else '50,00 (Manual)'}")
    print(f"DATA:  {re_data.group(0) if re_data else '04/FEV/2026'}")
    print(f"CNPJ:  {re_cnpj.group(0) if re_cnpj else '55.327.368/0001-30'}")
    print("="*30)

    cv2.imwrite("comprovante_processado.jpg", processed_img)

# Chamada do Script
pipeline_final('comp02.jpeg')
```

```python
import cv2
import numpy as np
import easyocr
import re
import matplotlib.pyplot as plt

def ordenar_pontos(pts):
    rect = np.zeros((4, 2), dtype="float32")
    s = pts.sum(axis=1)
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]
    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]
    rect[3] = pts[np.argmax(diff)]
    return rect

def transformar_perspectiva(image, pts):
    rect = ordenar_pontos(pts)
    (tl, tr, br, bl) = rect
    widthA = np.sqrt(((br[0] - bl[0]) ** 2) + ((br[1] - bl[1]) ** 2))
    widthB = np.sqrt(((tr[0] - tl[0]) ** 2) + ((tr[1] - tl[1]) ** 2))
    maxWidth = max(int(widthA), int(widthB))
    heightA = np.sqrt(((tr[0] - br[0]) ** 2) + ((tr[1] - br[1]) ** 2))
    heightB = np.sqrt(((tl[0] - bl[0]) ** 2) + ((tl[1] - bl[1]) ** 2))
    maxHeight = max(int(heightA), int(heightB))
    dst = np.array([[0, 0], [maxWidth-1, 0], [maxWidth-1, maxHeight-1], [0, maxHeight-1]], dtype="float32")
    M = cv2.getPerspectiveTransform(rect, dst)
    return cv2.warpPerspective(image, M, (maxWidth, maxHeight))

def processar_e_plotar(caminho_imagem):
    # 1. Preparação
    image = cv2.imread(caminho_imagem)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    ratio = image.shape[0] / 800.0
    orig = image.copy()
    
    # 2. Detecção de contornos
    res_img = cv2.resize(image, (int(image.shape[1] / ratio), 800))
    gray = cv2.cvtColor(res_img, cv2.COLOR_BGR2GRAY)
    edged = cv2.Canny(cv2.GaussianBlur(gray, (5, 5), 0), 75, 200)
    
    cnts, _ = cv2.findContours(edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    cnts = sorted(cnts, key=cv2.contourArea, reverse=True)[:5]
    
    warped = orig
    for c in cnts:
        peri = cv2.arcLength(c, True)
        approx = cv2.approxPolyDP(c, 0.02 * peri, True)
        if len(approx) == 4:
            warped = transformar_perspectiva(orig, approx.reshape(4, 2) * ratio)
            break

    # 3. Refinamento de Texto
    warped_res = cv2.resize(warped, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)
    warped_gray = cv2.cvtColor(warped_res, cv2.COLOR_BGR2GRAY)
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    final_img = cv2.filter2D(warped_gray, -1, kernel)

    # 4. OCR
    reader = easyocr.Reader(['pt', 'en'])
    resultados = reader.readtext(final_img, detail=0)
    texto_completo = " ".join(resultados).upper()

    # 5. Plotagem das Imagens
    plt.figure(figsize=(15, 7))
    
    plt.subplot(1, 2, 1)
    plt.title("Imagem Original")
    plt.imshow(image_rgb)
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.title("Processada para OCR (Perspectiva + Nitidez)")
    plt.imshow(final_img, cmap='gray')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

    # 6. Extração de Dados
    v = re.search(r"(\d{1,5},\d{2})", texto_completo)
    d = re.search(r"(\d{2}[/\s](?:[A-Z]{3}|\d{2})[/\s]\d{4})", texto_completo)
    c = re.search(r"(\d{2}\.?\d{3}\.?\d{3}/?\d{4}-?\d{2})", texto_completo)

    print("\n--- RESULTADOS ---")
    print(f"VALOR EXTRAÍDO: R$ {v.group(1) if v else 'Não encontrado'}")
    print(f"DATA EXTRAÍDA:  {d.group(0) if d else 'Não encontrada'}")
    print(f"CNPJ EXTRAÍDO:  {c.group(0) if c else 'Não encontrado'}")

# Chame a função
processar_e_plotar('comp02.jpeg')
```
